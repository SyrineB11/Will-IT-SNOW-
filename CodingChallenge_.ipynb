````python
#Transform boolean columns from True/False to  1/0
boolean_columns = ['fog', 'rain', 'snow', 'hail', 'thunder', 'tornado']
Data_task2[boolean_columns] = Data_task2[boolean_columns].astype(int)
````

````python
Data_task2.head(2)
````

````python
def get_features_for_date(date):
    'This function will give us all the relevent features for a given date'

    data_for_date = Data_task2[Data_task2['date'] == date]
    relevant_features = list(Data_task2.columns)
    relevant_features.remove('date')
    relevant_features.remove('snow')

    X_features = data_for_date[relevant_features].copy()

    return X_features
````

````python
import xgboost as xgb
from sklearn.metrics import accuracy_score
````

````python
X_train, y_train = train_set.drop(columns=['snow', 'date']), train_set['snow']
X_eval, y_eval = eval_set.drop(columns=['snow', 'date']), eval_set['snow']
X_test, y_test = test_set.drop(columns=['snow', 'date']), test_set['snow']
````

````python
xgb_model = xgb.XGBClassifier()
xgb_model.fit(X_train, y_train)
````

````python
xgb_eval_pred = xgb_model.predict(X_eval)
xgb_eval_accuracy = accuracy_score(y_eval, xgb_eval_pred)
print('XGBoost Evaluation Accuracy:', xgb_eval_accuracy)
````

````python
xgb_test_pred = xgb_model.predict(X_test)
xgb_test_accuracy = accuracy_score(y_test, xgb_test_pred)
print('XGBoost Test Accuracy:', xgb_test_accuracy)
````

````python
tomorrow_date=dt.datetime.today()+timedelta(days=1)
tomorrow_date.date().isoformat()
````

````python
def predict_snow():
   '''This function will predict in realtime if tomorrow 15 years ago will snow or not'''
   today = dt.datetime.today()
   tomorrow = today + timedelta(days=1)
   fifteen_years_ago = tomorrow - timedelta(days=int(15 * 365))
   fifteen_years_ago_str = str(fifteen_years_ago).split(' ')[0]
   X_features = get_features_for_date(fifteen_years_ago_str)
   result= xgb_model.predict(X_features)
   if result[0]==0:
    print(f"No snow tomorrow {fifteen_years_ago_str} ! you can still enjoy your day :D")
   else:
    print(f"Yayy! It will snow tomorrow the {fifteen_years_ago_str}. Wear a nice coat a,d stay warm!")
````
